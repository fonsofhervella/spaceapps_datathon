```python
#subplots

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns



# Time-based analysis
df['datetime_DSCVR'] = pd.to_datetime(df['datetime_DSCVR'])
df.set_index('datetime_DSCVR', inplace=True)

# Create a figure with two subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Plot Kp data on the first subplot
axs[0].plot(df.index, df['Kp'], label='Kp', color='blue')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Planetary K Index Mean')
axs[0].set_title('Kp Time Series')

# Plot ap data on the second subplot
axs[1].plot(df.index, df['ap'], label='ap', color='red')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('ap')
axs[1].set_title('ap Time Series')

# Allow overlap of x-axis labels on both subplots
for ax in axs:
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True)

# Fill between the two lines with transparency
axs[0].fill_between(df.index, df['Kp'], df['ap'], alpha=0.2, color='gray', label='Overlap')
axs[1].fill_between(df.index, df['ap'], df['Kp'], alpha=0.2, color='gray')

# Display legends for both subplots
axs[0].legend()
axs[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Pairplot (scatter plots)
sns.pairplot(df[['MagFieldVect_Comp1_mean', 'MagFieldVect_Comp2_mean', 'MagFieldVect_Comp3_mean']])
plt.suptitle('Pairplot of Magnetic Field Components', y=1.02)
plt.show()
```

```python
# Distribution plots
plt.figure(figsize=(12, 6))
sns.histplot(data=df[['FarCup1_mean', 'FarCup2_mean', 'FarCup3_mean']])
plt.title('Distribution of Faraday Cup Measurements')
plt.xlabel('Measurement Values')
plt.show()
```

```python
#do a double plot of both ap and kp to see if there is a correlation side by side 
#plotting the data
plt.figure(figsize=(10,5))
plt.plot(df['datetime_DSCVR'], df['Kp'])
plt.plot(df['datetime_DSCVR'], df['ap'])
#allow overlap of x-axis labels
plt.xticks(rotation=45)
#se transparency to see both plots
plt.fill_between(df['datetime_DSCVR'], df['Kp'], df['ap'], alpha=0.2)

plt.xlabel('Time')
plt.ylabel('Planetary K Index Mean')
plt.show()
 
```

```python
# Plot time series data of magnetic field components putting datetime on x-axis and mean value of each component on y-axis
plt.figure(figsize=(14, 6))
plt.plot(df['datetime_DSCVR'], df['MagFieldVect_Comp1_mean'], label='MagFieldVect_Comp1_mean')
plt.plot(df['datetime_DSCVR'], df['MagFieldVect_Comp2_mean'], label='MagFieldVect_Comp2_mean')
plt.plot(df['datetime_DSCVR'], df['MagFieldVect_Comp3_mean'], label='MagFieldVect_Comp3_mean')
plt.title('Time Series of Magnetic Field Components')
plt.xlabel('Date')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
